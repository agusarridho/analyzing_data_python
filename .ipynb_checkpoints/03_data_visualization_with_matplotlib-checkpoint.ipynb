# Matplotlib

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
from numpy.random import randn # for generating random number

# so that plots appear immediately in the Notebook
%matplotlib inline
```

### Scatter Plot

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
dframe_wine = pd.read_csv(url,sep=';')
dframe_wine.plot(kind='scatter',x='quality',y='alcohol')
```

### Bar Plot with Error Bars

```python
N=5
ind = np.arange(N) # the x locations for the groups width = 0.35 
width = 0.35 # the width of the bars
fig, ax = plt.subplots()

menMeans = (20, 35, 30, 35, 27)
menStd = (2, 3, 4, 1, 2)
rects1 = ax.bar(ind, menMeans, width, color='r', yerr=menStd)
womenMeans = (25, 32, 34, 20, 25)
womenStd = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, womenMeans, width, color='y', yerr=womenStd)

# add axis labels, title and axis tickmarks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend((rects1[0], rects2[0]), ('Men','Women'))

# you can add data labels above each bar
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,'%d' % int(height),ha='center', va='bottom')
        
autolabel(rects1)
autolabel(rects2)
```

### 3D Graphical Analysis

```python
# grab an array of 1000 datapoints
points = np.arange(-5,5,0.01)
# create the grid
dx,dy=np.meshgrid(points,points)
# set an evaluating function
z = (np.sin(dx) + np.sin(dy))

# plot the array - NOTE: plots the positions 1-1000, not the values -5 to 5
plt.imshow(z)
# add a colorbar & title
plt.colorbar()
# add a chart title
plt.title("Plot for sin(x)+sin(y)")
```

### Histogram

```python
# data grouped into 10 bins by default,
# with 11 equally spaced borders (min to max)
dataset1 = randn(100)
plt.hist(dataset1)
```

```python
# dataset2 is set to indianred for clarity
dataset2 = randn(80)
plt.hist(dataset2,color='indianred')
```

```python
# plot both histograms in the same Jupyter notebook cell
# normed=True normalizes the data (since we have two different-sized datasets)
# alpha=0.5 sets the transparency
plt.hist(dataset1,normed=True,alpha=0.5,bins=20) 
plt.hist(dataset2,normed=True,color='indianred',alpha=0.5,bins=20)
```
