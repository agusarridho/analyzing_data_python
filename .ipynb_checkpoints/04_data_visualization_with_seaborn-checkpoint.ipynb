# Seaborn

### Import Libraries

```python
import numpy as np
import pandas as pd
from numpy.random import randn # for generating random number
from scipy import stats # numpy stats library
import matplotlib as mpl # plotting  modules and libraries
import matplotlib.pyplot as plt
import seaborn as sns

# so that plots appear immediately in the Notebook
%matplotlib inline
```

### Rug Plots

```python
# plots a simple row of tic marks along the x-axis
dataset = randn(25)
sns.rugplot(dataset)
```

### Histogram using Factorplot

Note: Histograms are already part of matplotlib: plt.hist(dataset)

Seaborn's factorplot lets you choose between histograms, point plots, violin plots, etc.

Also, the "hue" argument makes it easy to compare multiple variables simultaneously.

Unfortunately, sorting columns appropriately can be a challenge.

The following example makes use of the Iris flower data set included in Seaborn:

```python
iris = sns.load_dataset("iris")
xorder = np.apply_along_axis(sorted, 0, iris['petal_length'].unique())
# Note: without size=8, the x-axis labels overlap
sns.factorplot('petal_length', data=iris, order=xorder, size=8, hue='species',kind='count');
```

### KDE Plots (Kernel Density Estimation Plots)

```python
dataset = randn(25)
sns.kdeplot(dataset)
```

```python
sns.kdeplot(dataset,cumulative=True)
```

### Combined Plots (kde, hist, rug) using Distplot

```python
# by default, a KDE over a histogram 
sns.distplot(dataset,bins=25)
# here a rug and a KDE
sns.distplot(dataset,rug=True,hist=False)
# To control specific plots in distplot, use a [plot]_kws argument with dictionaries: 
sns.distplot(dataset,bins=25,kde_kws={'color':'indianred','label':'KDE PLOT'}, 
             hist_kws={'color':'blue','label':"HISTOGRAM"})
```

### Box and Whisker Plots

```python
# Box plots are another tool for representing Probability Density Functions
data1 = randn(100)
data2 = randn(100)
sns.boxplot(data=[data1,data2])
```

### Violin Plots

```python
# May reveal what a box plot doesn't by incorporating 
# some of the functionality of KDE plots
sns.violinplot(data=[data1,data2])
```

### Joint Plots

```python
data1 = randn(1000)
data2 = randn(1000)
sns.jointplot(data1,data2)
```

```python
sns.jointplot(data1,data2,kind='hex')
```

### Regression Plots

```python
tips = sns.load_dataset("tips")
sns.lmplot("total_bill","tip",tips)
```

```python
sns.lmplot("total_bill","tip",tips,order=4, scatter_kws={"color":"indianred"}, line_kws={"linewidth": 1, "color": "blue"})
```

### Heatmaps

```python
flight_dframe = sns.load_dataset('flights')
# Pivot the data to make it more usable (index=month, columns=year, fill=passengers): 
flight_dframe = flight_dframe.pivot("month","year","passengers")
sns.heatmap(flight_dframe)
```

```python
# You can add fill data with
sns.heatmap(flight_dframe,annot=True,fmt='d')
```

```python
# You can specify a "center" for the colormap with 
sns.heatmap(flight_dframe,center=flight_dframe.loc['January',1955])
```

### Clustered Matrices

```python
cg = sns.clustermap(flight_dframe)
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
```
